### Data Frames

##### To manually store data in a table, create a DataFrame

#### A DataFrame is a 2-dimensional data structure that can store data of different types (including characters, integers, floating point values, categorical data and more) in columns. It is similar to a spreadsheet.

```python
df = pd.DataFrame(
    {
        "Name" : [
            "Mr.Punjan Pandey",
            "Mr.Alex Huggs",
            "Mr.Virat Kohli",
            "Mr.Henry Ford"
        ],
        "Age":[18,59,35,40],
        "Sex":["male","male","male","male"],

    }
)
df
```
