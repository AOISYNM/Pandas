### Just doing air2.plot.area() shows all the dara wihtout fixed height and all 

```python
axs = air2.plot.area()
```

```python
fig, axs = plt.subplots(figsize=(12, 4))
air2.plot.area(ax=axs, alpha=0.7)  # transparency

axs.set_ylabel("NO$_2$ concentration")
axs.set_xlabel("Date")  # custom x-label
axs.set_title("Air Quality Over Time", fontsize=14)
axs.grid(True, alpha=0.3)  # light gridlines

fig.tight_layout()  # prevents label cutoff
fig.savefig("no2_concentrations.png", dpi=300, bbox_inches='tight')

plt.show()
```

#### plt.subplots() = creates a new figure with axes
#### fig = the entire figure (like a canvas/paper)
#### axs = the axes (the actual plot area where data is drawn)
#### figsize=(12, 4) = canvas size: 12 inches wide, 4 inches tall
#### .plot.area() = creates a stacked area chart (filled line plot)
#### ax=axs = "draw on this specific axes object"
#### alpha=0.7 = 70% opacity (0=invisible, 1=solid). Makes overlapping areas visible
#### Adds text label to y-axis (vertical axis)
#### $_2$ = LaTeX syntax that renders as subscript: NO₂
#### axs.set_xlabel("Date")

#### Adds text label to x-axis (horizontal axis)
#### axs.set_title("Air Quality Over Time", fontsize=14)

#### Adds title at the top of the plot
#### fontsize=14 = makes title text 14 points (bigger than default)
#### axs.grid(True, alpha=0.3)

#### True = turn on gridlines
#### alpha=0.3 = 30% opacity (light, subtle lines that don't overpower the data)
#### fig.tight_layout()

#### Automatically adjusts spacing so labels don't get cut off
#### Prevents xlabel/ylabel/title from being cropped at edges
#### "no2_concentrations.png" = filename to save
#### dpi=300 = dots per inch (higher = better quality, 300 is print-quality)
#### bbox_inches='tight' = crop extra whitespace around the plot
