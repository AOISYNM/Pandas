# Concatenate columns in pandas 

```python
import pandas as pd
```

```python
airno2 = pd.read_csv("../Data/air_quality_no2_long.csv")
airno2 = airno2[["date.utc", "location","parameter", "value"]]
airno2.head()
```

```python
airpm25 = pd.read_csv("air_quality_pm25_long.csv")
airpm25 = airpm25[["date.utc", "location","parameter", "value"]]
airpm25.head()
```

```python
air = pd.concat([airno2,airpm25],axis=0)
air
```

#### The concat() function performs concatenation operations of multiple tables along one of the axes (row-wise or column-wise).

By default concatenation is along axis 0, so the resulting table combines the rows of the input tables.

```python
air.sort_values("date.utc")
```
